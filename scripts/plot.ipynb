```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.set(rc={'figure.figsize':(11,8)})
sns.set(font_scale=1.5)
```

```python
def readData(path,delimit):
    df = pd.read_csv(path,delimiter=delimit)
    df = df[['processes','solution time [s]']]
    plot_df = df.groupby('processes').mean()
    return plot_df, df
```

```python
#MPI datensätze
dn9 = readData('../rawdata/new data/mpi/sudokuNormal9_27082022222135.csv',';')

dh9 = readData('../rawdata/new data/mpi/sudokuHard9_28082022002212.csv',';')

dv9 = readData('../rawdata/new data/mpi/sudokuVeryHard9_27082022224426.csv',',')

dn16 = readData('../rawdata/new data/mpi/sudokuNormal16_27082022222437.csv',',')
```

```python
#OMP datensätze
#naiv

odn9 = readData('../rawdata/new data/omp_naive/sudokuNormal9_28082022051501.csv',';')

odh9 = readData('../rawdata/new data/omp_naive/sudokuHard9_28082022054748.csv',';')

odv9 = readData('../rawdata/new data/omp_naive/sudokuVeryHard9_28082022054811.csv',';')

odn16 = readData('../rawdata/new data/omp_naive/sudokuNormal16_28082022060243.csv',';')
#task

otdn9 = readData('../rawdata/new data/omp_task/sudokuNormal9_28082022153301.csv',';')

otdh9 = readData('../rawdata/new data/omp_task/sudokuHard9_28082022153308.csv',';')

otdv9 = readData('../rawdata/new data/omp_task/sudokuVeryHard9_28082022153313.csv',';')

otdn16 = readData('../rawdata/new data/omp_task/sudokuNormal16_28082022162629.csv',';')

```

```python
ax = dn9[0].plot()
odn9[0].plot(ax=ax)
otdn9[0].plot(ax=ax)
ax.legend(["MPI","OMP naive","OMP task"])
ax.set_ylabel('Lösungszeit [s]')
ax.set(xlabel='Anzahl Cores')
ax.set(title='9x9 Sudoku Normal')
```
